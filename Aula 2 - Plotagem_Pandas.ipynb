## Exercício 1:
Utilizando a base de dados de covid:

**a)** Faça um gráfico de barras mostrando o total de óbitos (considerando todos os meses) para todos os DRS, excluindo a Grande São Paulo. Coloque título no gráfico e nos eixos, defina um tamanho de imagem bom e intercale as barras com as cores 'red' e 'green'.

```python
df2_aux =  df2[df2['nome_drs'] != 'Grande São Paulo']
df2_aux['nome_drs'].unique()
```

```python
df2_aux.groupby(['nome_drs'], as_index=False).sum().plot(x = 'nome_drs', 
                                                         y = 'obitos_novos',
                                                         kind = 'bar',
                                                        color = ['red', 'green'], 
                                                        title = 'Número de óbitos por DRS',
                                                        legend = False,
                                                        xlabel='Nome DRS', 
                                                        ylabel='Total de óbitos',
                                                        rot= 30,
                                                        figsize = (16, 10));
```

**b)**  Faça um gráfico de barras horizontal mostrando o total de óbitos no mês de abril para 3 DRS da sua escolha. Coloque título no gráfico e nos eixos, defina um tamanho de imagem bom.

```python
tres_drs = ['Marília', 'Barretos', 'Piracicaba']
df2_april = df2[df2['nome_drs'].isin(tres_drs)
                & (df2.ano_mes=='2020-04')].groupby(['nome_drs'], as_index= False).sum()
df2_april
```

```python
df2_april.plot(x = 'nome_drs', 
                y = 'obitos_novos',
                kind = 'barh',
                color = ['red', 'green', 'blue'], 
                title = 'Número de óbitos por DRS',
                legend = False,
                xlabel='Nome DRS', 
                ylabel='Total de óbitos',
                figsize = (16, 10));
```

## Exercício 2

**a)** Utilizando a base de dados da IBM, crie um histograma para as idades dos funcionários, separando por gênero.

```python
ax_1 = df1[df1['Gender']=='Male'].plot(y='Age', kind='hist', 
                                          title= 'Idade dos funcionários da IBM',
                                          figsize=(15,10),
                                          bins = 12, 
                                            label='Masculino', color= 'blue')

ax_2 = df1[df1['Gender']=='Female'].plot(y='Age', kind='hist',
                                             figsize=(15,10), bins= 12, label='Feminino', 
                                            color='red',  ax=ax_1)

ax_1.set_xlabel('Idade', fontsize=15)
ax_1.set_ylabel('Número de funcionários', fontsize=15)

ax_1.plot();
```
